```python
x=Xnew.drop(columns='listed_price')
y=Xnew['listed_price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['transmission','fuel','dvn','oem','model','variant','City','Color','No of Cylinder','Gear Box','Seats','Steering Type','Front Brake Type','Rear Brake Type','Tyre Type','state','exterior_color','owner_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['transmission','fuel','dvn','oem','model','variant','City','Color','No of Cylinder','Gear Box','Seats','Steering Type','Front Brake Type','Rear Brake Type','Tyre Type','state','exterior_color','owner_type']),
                                     remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
Y_pred
```

```python
print("Accuracy is",r2_score(Y_test,Y_pred))
```

For 70:30 ratio


```python
scores=[]
for i in range(1000):
  X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=i)
  lr=LinearRegression()
  pipe=make_pipeline(column_trans,lr)
  pipe.fit(X_train,Y_train)
  Y_pred=pipe.predict(X_test)
  scores.append(r2_score(Y_test,Y_pred))

```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
Y_pred=pipe.predict(X_test)
print("Accuracy is",r2_score(Y_test,Y_pred))
```
